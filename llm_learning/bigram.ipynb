```python
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
with open('wizard_of_oz.txt', 'r', encoding='utf-8') as f:
    text=f.read()
chars = sorted(set(text))
print(chars)
vocabulary_size = len(chars)
print(vocabulary_size)
```

```python
string_to_int = { ch:i for i,ch in enumerate(chars) }
int_to_string = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [string_to_int[c] for c in s]
decode = lambda l: ''.join([int_to_string[c] for c in l])

mystery = encode('hello')
print(mystery)
print(decode(mystery))

data = torch.tensor(encode(text), dtype=torch.long)
print(data[:100])
```

```python
# here we specify that the data created above is split into 80-20 split
# n is the character at the 80% mark of the data
n = int(0.8 * len(data))

train_data = data[:n]
val_data = data[n:]
```

```python
block_size = 8

x = train_data[:block_size]
y = train_data[1:block_size+1]

for t in range(block_size):
    context = x[:t+1]
    target = y[t]
    print('when input is', context, 'target is', target)
```
